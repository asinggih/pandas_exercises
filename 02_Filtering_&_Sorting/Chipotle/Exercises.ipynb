### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"

chipo = pd.read_csv(url, sep="\t")
```

### Step 4. How many products cost more than $10.00?

```python
# clean up prices to only hold value, no "$" sign, and convert them to flat
# prices = [float(price[1:len(price)]) for price in chipo.item_price]

# chipo.item_price = prices
# chipo.item_price

# dropping dupes in that has the same name and quantity
chipo_cleaned = chipo.drop_duplicates(['item_name', 'quantity'], keep="first", inplace=False)

# "The Where condition"
single_quant = chipo_cleaned.quantity == 1
more_than_ten = chipo_cleaned.item_price > 10

chipo_cleaned[single_quant & more_than_ten].item_name.nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
single_quantity = chipo_cleaned[single_quant][['item_name', 'item_price']]
single_quantity.sort_values(by="item_price", ascending=False)
```

### Step 6. Sort by the name of the item

```python
chipo_cleaned.sort_values(by=['item_name'])
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo_cleaned.sort_values(by=['item_price'], ascending=False).head(1)
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python

veg_sld_bwl = chipo.item_name == 'Veggie Salad Bowl'
len(chipo[veg_sld_bwl])
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
canned_soda = chipo.item_name == 'Canned Soda'
extra = chipo.quantity > 1
len(chipo[canned_soda & extra])





```
